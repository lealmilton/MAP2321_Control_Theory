# Introduction

An inverted pendulum is a pendulum that has its center of mass above its pivot point. It is unstable and without additional help will fall over. It can be suspended stably in this inverted position by using a control system to monitor the angle of the pole and move the pivot point horizontally back under the center of mass when it starts to fall over, keeping it balanced. The inverted pendulum is a classic problem in dynamics and control theory and is used as a benchmark for testing control strategies. It is often implemented with the pivot point mounted on a cart that can move horizontally under control of an electronic servo system as shown in the photo; this is called a cart and pole apparatus.

Source: [Inverted Pendulum](https://en.wikipedia.org/wiki/Inverted_pendulum)

# Motion Equations

$
\begin{cases}
(m + M)\ddot{z} + m l \cos(\theta)\ddot{\theta} - m l \sin(\theta)\dot{\theta}^2 = u \\
m l \ddot{\theta} + m\cos(\theta)\ddot{z} - m g \sin(\theta) = 0
\end{cases}
$

*The derivation of the equations is elaborated in the work of Ogata, K. (2010). Modern Control Engineering (5th ed.). Prentice Hall.*

If we consider $u=0$ and $\dot{\theta} = \ddot{\theta} = \dot{z} = \ddot{z} = 0 $, then:

$
\begin{cases}
(m + M)0 + m l \cos(0)0 - \mu l \sin(0)0 = 0 \Rightarrow 0 = 0 \\
m l 0 + m\cos(0)0 - m g \sin(0) = 0 \Rightarrow \sin(0) = 0 \Leftrightarrow \theta = 0 \text{ or } \theta = \pi
\end{cases}
$

Note: $z$ can be an arbitrary value.

We are interested in the equilibrium:

$
\begin{cases}
\theta = \bar{\theta} = 0 \\
z = \bar{z} = 0
\end{cases}
$

To analyze the dynamics close to $(\bar{\theta}, \bar{z})$ we will linearize the system:

$
\begin{cases}
\theta = \bar{\theta} + \delta_{\theta} \\
z = \bar{z} + \delta_{ z}
\end{cases}
$

Note that as $\bar{\theta} = 0 $ and $ \bar{z} = 0$, we have:

$
\begin{cases}
\theta =  \delta_{\theta} \\
z = \delta_{ z}
\end{cases}
$

We substitute the linearized equations into the original equations and consider the Taylor expansion for the sine and cosine functions.

$
\begin{cases}
(m+M)\ddot{\delta_{ z}} + m l \left(1 - \frac{(\delta_{ \theta})^2}{2}\right)\ddot{\delta_{\theta}} - m l \left(\delta_{\theta} - \frac{(\delta_{ \theta})^3}{6}\right)\dot{\delta_{ \theta}}^2 = u, \\
m l \ddot{\delta_{ \theta}} + m\left(1 - \frac{(\delta_{ \theta})^2}{2}\right)\ddot{\delta_{z}}- m g \left(\delta_{\theta} - \frac{(\delta_{ \theta})^3}{6}\right) = 0.
\end{cases}
$


 If we neglect all terms of higher order than $\delta^2$, since $\delta$ is small, and simplify the notation as follows:

 $
\begin{cases}
\delta_{\theta} \rightarrow \theta \\
\delta_{z} \rightarrow z
\end{cases}
$


Then, we get:

$
\begin{cases}
(m+M)\ddot{z} + m l \ddot{\theta} &= u \quad &\text{(a)} \\
m l \ddot{\theta} + m\ddot{z} - m g \theta &= 0 \quad &\text{(b)}
\end{cases}
$


If we compute $ (a) - (b) $ and then compute $ (b)\dfrac{(m+M)}{m} - (a)$, we end up with the final system:

$
\begin{cases}
M\ddot{z} + m g \theta = u \\
lM\ddot{\theta} - (m+M)g\theta = -u
\end{cases}
$


# State Variables

Let's define the state variables as follows:

$
\begin{cases}
x_1 &= \theta \\
x_2 &= \dot{\theta} \\
x_3 &= z \\
x_4 &= \dot{z}
\end{cases}
$

Deriving the state variables and substituting, gives us:



$
\begin{cases}
\dot{x}_1 = \dot{\theta} = x_2 \\
\dot{x}_2 = \ddot{\theta} = \frac{(m+M)g}{Ml} x_1 - \frac{1}{Ml} u \\
\dot{x}_3 = \dot{z} = x_4 \\
\dot{x}_4 = \ddot{z} = -\frac{m g}{M} x_1 + \frac{1}{M} u
\end{cases}
$


We can write the system in matrix form:

$
\begin{aligned}
\begin{bmatrix}
\dot{x}_1 \\
\dot{x}_2 \\
\dot{x}_3 \\
\dot{x}_4
\end{bmatrix} &=
\underbrace{
\begin{bmatrix}
0 & 1 & 0 & 0 \\
\frac{(m+M)g}{Ml} & 0 & 0 & 0 \\
0 & 0 & 0 & 1 \\
-\frac{mg}{M} & 0 & 0 & 0
\end{bmatrix}
}_{A}
\begin{bmatrix}
x_1 \\
x_2 \\
x_3 \\
x_4
\end{bmatrix} +
\underbrace{
\begin{bmatrix}
0 \\
-\frac{1}{Ml} \\
0 \\
\frac{1}{M}
\end{bmatrix}
}_{B} u
\end{aligned}
$


The system output that we are interested in will be the position of the cart, given by $z= x_3$ and obtained as follows:

$
y = \begin{pmatrix}
0 & 0 & 1 & 0
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2 \\
x_3 \\
x_4
\end{pmatrix}
$

### a) Verify the controllability of the system $ \dot{x} = Ax + Bu $ using the Kalman Rank condition.

The controllability matrix $C_r$ is given by:

$
\mathcal{C_r} = \left[\begin{array}{c:c:c:c}
B & AB & A^2B & A^3B
\end{array}\right]
$

The system is controllable if and only if the rank $R$ of  $C_r$ is full. In our case, we should get $R=4$.


```python
import sympy as sp
from IPython.display import display, Math

# Initialize pretty printing
sp.init_printing(use_unicode=True, use_latex='mathjax')

# Define symbols
m, M, g, l = sp.symbols('m M g l')

# Define matrices A and B
A = sp.Matrix([
    [0, 1, 0, 0],
    [((m + M)/M)*(g/l), 0, 0, 0],
    [0, 0, 0, 1],
    [(-m*g)/M, 0, 0, 0]
])
B = sp.Matrix([0, (-1)/(M*l), 0, 1/M])

# Initialize controlability matrix
controllability_matrix = B
for i in range(1, A.shape[0]):
    controllability_matrix = controllability_matrix.row_join(A**i * B)

# Display the controllability matrix in a larger format using '\Huge'
display(Math(r'\text{{Controllability Matrix }}:'))
print('\n')
display(Math(r'\mathcal{C}_r = \LARGE ' + sp.latex(controllability_matrix)))
print('\n')

# Display the rank of the controllability matrix in a larger format
rank = controllability_matrix.rank()
display(Math(r'\text{{Rank of }} \mathcal{{C_r}} : '  + sp.latex(rank)))
print('\n')

# Display if the system is controllable in a larger format
system_controllable = 'controllable' if rank == A.shape[0] else 'not controllable'
display(Math(r'\text{{The system is }}'  + system_controllable))
```

### b) Verify the observability of the system

$
\begin{cases}
\dot{x} = Ax + Bu \\
y = Cx
\end{cases}
$

using the Kalman Rank condition.

The observability matrix $O_b$ is given by:

$
\mathcal{O_b} = \left[\begin{array}{c:c:c:c}
C^T & A^TC^T & (A^2)^TC^T & (A^3)^TC^T
\end{array}\right]
$

The system is controllable if and only if the rank $R$ of  $O_b$ is full. In our case, we should get $R=4$.


```python
# Define matrix C
C = sp.Matrix([[0, 0, 1, 0]])

# Initialize observability matrix
observability_matrix = C.transpose()
for i in range(1, A.shape[0]):
    observability_matrix = observability_matrix.row_join((A.transpose()**i * C.transpose()))

# Display the observability matrix in a larger format using '\Huge'
display(Math(r'\text{{Observability Matrix }}:'))
print('\n')
display(Math(r'\mathcal{O}_b = \LARGE ' + sp.latex(observability_matrix)))
print('\n')

# Display the rank of the observability matrix in a larger format
rank = observability_matrix.rank()
display(Math(r'\text{{Rank of }} \mathcal{{O_b}} : '  + sp.latex(rank)))
print('\n')

# Display if the system is observable in a larger format
system_observable = 'observable' if rank == A.shape[0] else 'not observable'
display(Math(r'\text{{The system is }}'  + system_observable))
```

### c) Construct the following matrix:

$
\mathcal{P} = \left[\begin{array}{c:c:c:c}
A^3B & A^2B & AB & B
\end{array}\right]
$

```python
# Compute the necessary powers of A and multiply by B
A3B = A**3 * B
A2B = A**2 * B
AB = A * B

# Concatenate the results to form the P matrix
P = A3B.row_join(A2B).row_join(AB).row_join(B)

# Display the P matrix in the same format as previously done

print('\n')
display(Math(r'\mathcal{P} = \LARGE ' + sp.latex(P)))
```

### d) Compute $P^{-1}$:

```python
# Check if P is invertible and calculate the inverse if it is
if P.det() != 0:
    P_inverse = P.inv()
    # Display the inverse of P in a larger format using '\Huge'
    display(Math(r'\text{{Inverse of Matrix }} P :'))
    print('\n')
    display(Math(r'\mathcal{P}^{-1} = \LARGE ' + sp.latex(P_inverse)))
else:
    print("Matrix P is not invertible.")
```

### e) Compute $A_1 = P^{-1}AP$ and $B_1=P^{-1}B$

```python
# Transform matrices A and B using P inverse
A_1 = sp.simplify(P_inverse * A * P)
B_1 = sp.simplify(P_inverse * B)

# Display transformed matrices A_1 and B_1
print('\n')
display(Math(r'\mathcal{A}_1 = \LARGE ' + sp.latex(A_1)))
print('\n')
display(Math(r'\mathcal{B}_1 = \LARGE ' + sp.latex(B_1)))
print('\n')

# Retrieve coefficients from the first column of A_1
coefs = A_1.col(0)

# Generate LaTeX code for each coefficient, to be displayed one below the other
coef_display = '\\\\ '.join([f'a_{{1{i+1}}} = {sp.latex(coef)}' for i, coef in enumerate(coefs)])

# Display the coefficients using Math to render as LaTeX
display(Math(r'\text{{Coefficients of the first column of }} A_1: \\ ' + coef_display))
```
